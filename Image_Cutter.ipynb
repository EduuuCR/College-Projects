```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
azul = (255,0,0)
verde = (0,255,0)
vermelho = (0,0,255)
fonte = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img1 = cv2.imread('img1.png')
cv2_imshow(img1)
```

```python
recortep = img1[200:320,280:340]
cv2_imshow(recortep)
recortec1 = img1[190:250,135:185]
cv2_imshow(recortec1)
recortec2 = img1[215:280,355:455]
cv2_imshow(recortec2)
```

```python
mascara1 = np.zeros(img1.shape[:2], dtype = "uint8")
(cX1,cY1) = (160,215)
(cX2,cY2) = (405,245)
(pX,pY) = (315,260)
cv2.circle(mascara1, (cX1,cY1), 40, 255, -1)
cv2.circle(mascara1, (cX2,cY2), 50, 255, -1)
cv2.circle(mascara1, (pX,pY), 60, 255, -1)
img1mascara = cv2.bitwise_and(img1, img1, mask = mascara1)
cv2_imshow(img1mascara)
```
